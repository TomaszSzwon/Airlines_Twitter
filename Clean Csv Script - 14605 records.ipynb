```python
import csv
import re

input_file = "Tweets_noisy.csv"
output_file = "Tweets_cleaned.csv"

with open(input_file, "r", encoding="utf-8") as infile, open(output_file, "w", encoding="utf-8", newline="") as outfile:
    lines = infile.readlines()
    
    cleaned_lines = []
    for line in lines:
        line = line.strip()

        # Usuń ";;;;;;" na końcu linii wraz z nową linią
        line = re.sub(r'";{6,}[\r\n]*$', '', line)

        # Usuń niepotrzebne cudzysłowy na początku linii
        line = re.sub(r'^"', '', line)

        # Usuń nadmiarowe średniki i cudzysłowy na końcu linii
        line = re.sub(r'[";]+$', '', line)
        
        # Napraw cudzysłowy w treści
        line = re.sub(r'""', '"', line) 
        line = re.sub(r'";', ';', line)
        line = re.sub(r';"', ';', line)  
      
    
        # Napraw błędne encje HTML
        line = re.sub(r'&amp";?', '&amp;', line)
        line = re.sub(r'&amp;"', '&amp;', line)
        line = re.sub(r'&gt"', '&gt', line)
        # Popraw literówki
        line = re.sub(r'\bhte\b', 'the', line)  # "hte" na "the"
        line = re.sub(r'\banohter\b', 'another', line)  # "anohter" na "another"
        line = re.sub(r'\bhtey\b', 'they', line)  # "htey" na "they"
        line = re.sub(r'null', 'Thank', line) # null na Thank
        line = re.sub(r'@USAirways I been sitting on this plane for 40 minutes', '"@USAirways I been sitting on this plane for 40 minutes', line) # null na Thank  
        line = re.sub(r'@united frankly worse customer service ever', '"@united frankly worse customer service ever', line) # null na Thank     
        line = re.sub(r'@SouthwestAir How to find out if my parents luggage', '"@SouthwestAir How to find out if my parents luggage', line) # null na Thank 
        line = re.sub(r'@united Beyond frustrated', '"@united Beyond frustrated', line) # null na Thank 
        line = re.sub(r'(?<=,)(\[)', r'"[', line)
        line = re.sub(r'(o])', r']"', line)
        line = re.sub(r'@united Can someone assist me with my DM', '"@united Can someone assist me with my DM', line) # null na Thank 
        cleaned_lines.append(line)
    # Zapisz oczyszczone linie do nowego pliku
    outfile.write("\n".join(cleaned_lines) + "\n")

print(f"Plik został oczyszczony i zapisany jako {output_file}")
```
